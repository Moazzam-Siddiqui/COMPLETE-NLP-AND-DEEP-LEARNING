```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent = ['the glass of milk',
        'the glass of juice',
        'the cup of tea',
        'I am a good boy',
        'I am a good developer',
        'understand the meaning of love',
        'your videos are good'
        ]
vocab_size = 1000
```

```python
one_hot_repr = [one_hot(words,vocab_size) for words in sent]
one_hot_repr
```

```python
#Word Embedding Representation

from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
import numpy as np
```

```python
"""we apply padding to let our sentences be remain of same length like some of them are 4 words while some are 5 that makes instability so
we choose to underdock with certain number like i choose 8 as the max length,here pre means we do the padding before the sentences"""
sent_length = 8
embedded_docs = pad_sequences(one_hot_repr,padding='pre',maxlen=sent_length)
embedded_docs
```

```python
#Feature Representation
dimension = 10
model = Sequential()
model.add(Embedding(vocab_size,dimension,input_length=sent_length))
model.build(input_shape=(None, sent_length))  # Explicitly build the model
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
