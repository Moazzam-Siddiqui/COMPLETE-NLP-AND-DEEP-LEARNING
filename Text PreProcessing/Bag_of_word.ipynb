```python
import pandas as pd
import re
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer

df = pd.read_csv('spam.csv', encoding='latin-1')

df = df[['v1', 'v2']]
df.columns = ['label', 'message']
print(df.head(5))

ps = PorterStemmer()
ls = WordNetLemmatizer()
corpus = []


for i in range(len(df)):
    review = re.sub(r'[^a-zA-Z]',' ', df['message'][i])  # Fix regex issue
    review = review.lower()
    review = review.split()
    review = [ls.lemmatize(word,pos='v') for word in review if word not in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)


corpus
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=10,binary=True) #Max features shows us top most frequent words like how we want to see top 2500,and binary will let us see only 1 or 0 instead of non binary showing us how many times it occured
X = cv.fit_transform(corpus).toarray()
```

```python
X
cv.vocabulary_
```
